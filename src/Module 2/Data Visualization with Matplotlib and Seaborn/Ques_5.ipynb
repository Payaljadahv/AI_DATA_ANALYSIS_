```python
# Basic Seaborn Scatter Plot
# Description: Using Seaborn, recreate the scatter plot from Question 2 for height vs. weight.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data setup
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Create a DataFrame for Seaborn
data = pd.DataFrame({'Height (cm)': heights, 'Weight (kg)': weights})

# Create the scatter plot
sns.scatterplot(x='Height (cm)', y='Weight (kg)', data=data, color='purple')

# Add title
plt.title('Height vs Weight')

# Show the plot
plt.show()
```
