```python
# Scatter Plot of Variable Relationships
# Description: Use Seaborn to create a scatter plot showing the relationship between two
# variables.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset('tips')

# Scatter plot of total_bill vs tip
sns.scatterplot(data=tips, x='total_bill', y='tip', color='dodgerblue')

# Add title and labels
plt.title('Scatter Plot: Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')

# Show plot
plt.show()
```
