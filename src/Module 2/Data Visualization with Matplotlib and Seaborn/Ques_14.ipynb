```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

# Customize font and style
plt.rcParams.update({
    'font.family': 'Arial',
    'font.size': 12,
    'axes.titlesize': 16,
    'axes.labelsize': 14
})

# Create the line plot
plt.plot(x, y, color='crimson', linestyle='--', marker='o', markersize=8, linewidth=2, markerfacecolor='yellow')

# Add title and labels
plt.title('Customized Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add grid
plt.grid(True, linestyle=':', linewidth=0.7)

# Show plot
plt.show()
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

# Customize font and style
plt.rcParams.update({
    'font.family': 'Arial',
    'font.size': 12,
    'axes.titlesize': 16,
    'axes.labelsize': 14
})

# Create the line plot
plt.plot(x, y, color='crimson', linestyle='--', marker='o', markersize=8, linewidth=2, markerfacecolor='yellow')

# Add title and labels
plt.title('Customized Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add grid
plt.grid(True, linestyle=':', linewidth=0.7)

# Show plot
plt.show()

```
