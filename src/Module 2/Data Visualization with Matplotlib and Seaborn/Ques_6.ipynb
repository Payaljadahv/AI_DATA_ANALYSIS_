```python
# Customizing Matplotlib Plot Styles
# Description: Customize the Matplotlib bar plot from Question 1 with a grid, different color, and custom fonts.
```

```python
import matplotlib.pyplot as plt

# Product names and sales data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Set a custom font and style
plt.rcParams.update({
    'font.family': 'serif',
    'font.size': 12,
    'axes.titlesize': 16,
    'axes.labelsize': 14
})

# Create the bar plot
plt.bar(products, sales, color='teal')

# Add grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add title and labels
plt.title('Product Sales (in thousands)')
plt.xlabel('Products')
plt.ylabel('Sales (in thousands)')

# Show the plot
plt.show()
```
