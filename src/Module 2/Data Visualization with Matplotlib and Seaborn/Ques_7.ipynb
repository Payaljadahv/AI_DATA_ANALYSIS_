```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data setup
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]
sizes = [10, 15, 13, 18, 12]  # Size data for point scaling

# Create a DataFrame
data = pd.DataFrame({
    'Height (cm)': heights,
    'Weight (kg)': weights,
    'Size': sizes
})

# Create the enhanced scatter plot
sns.scatterplot(
    x='Height (cm)',
    y='Weight (kg)',
    size='Size',
    data=data,
    legend=False,
    color='darkorange',
    sizes=(50, 300)  # Scale the bubble sizes
)

# Add title
plt.title('Height vs Weight (Point Size by Data)')

# Show the plot
plt.show()
```
