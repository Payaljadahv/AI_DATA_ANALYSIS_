```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
data = np.random.normal(loc=0, scale=1, size=1000)

# Create subplots: 2 rows, 1 column
fig, axes = plt.subplots(2, 1, figsize=(8, 8))

# Line plot on the first subplot
axes[0].plot(x, y, color='blue', linestyle='-', linewidth=2)
axes[0].set_title('Sine Wave')
axes[0].set_xlabel('X')
axes[0].set_ylabel('sin(X)')
axes[0].grid(True)

# Histogram on the second subplot
axes[1].hist(data, bins=30, color='orange', edgecolor='black', alpha=0.7)
axes[1].set_title('Histogram of Normal Distribution')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')
axes[1].grid(axis='y', linestyle='--')

# Adjust layout to prevent overlap
plt.tight_layout()

plt.show()

```
