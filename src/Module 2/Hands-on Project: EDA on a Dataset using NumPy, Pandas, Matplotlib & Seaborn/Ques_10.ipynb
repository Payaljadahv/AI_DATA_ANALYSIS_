```python
# Step 1: Set Up the Environment
# Objective: Ensure you have the right tools and packages installed.

# 1. Install the required libraries (run in terminal or Jupyter with !)
# !pip install numpy pandas matplotlib seaborn

# 2. Verify the installation by importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully.")



# Step 2: Load & Explore the Dataset
# Objective: Load data into a pandas DataFrame and obtain a basic understanding of its structure.

# 3. Load a CSV file into a DataFrame (using seaborn's iris dataset as example)
iris = sns.load_dataset('iris')  # replace with pd.read_csv('your_file.csv') if needed

# 4. Display the first few records to understand the structure
print("\nFirst 5 rows of the dataset:")
print(iris.head())

# 5. Get a summary of the dataset
print("\nDataset info:")
print(iris.info())

print("\nStatistical summary:")
print(iris.describe())



# Step 3: Perform NumPy Operations
# Objective: Utilize NumPy for basic numerical operations and array manipulations.

# 6. Convert a DataFrame column to a NumPy array and perform array operations like mean and sum.
sepal_length_np = iris['sepal_length'].to_numpy()
print("\nMean sepal length:", np.mean(sepal_length_np))
print("Sum of sepal length:", np.sum(sepal_length_np))

# 7. Create a NumPy array and calculate the variance and standard deviation.
arr = np.array([10, 20, 30, 40, 50])
print("\nArray:", arr)
print("Variance:", np.var(arr))
print("Standard deviation:", np.std(arr))

# 8. Use NumPy to filter based on conditions.
filtered_arr = arr[arr > 25]
print("Filtered array (values > 25):", filtered_arr)



# Step 4: Data Manipulation with Pandas
# Objective: Use Pandas to clean and manipulate dataset for analysis.

# 9. Handle missing data by filling or dropping.
# Introduce missing values for demo
iris_missing = iris.copy()
iris_missing.loc[::10, 'sepal_length'] = np.nan
print("\nMissing values before filling:")
print(iris_missing.isnull().sum())

# Fill missing values with median
iris_missing['sepal_length'].fillna(iris_missing['sepal_length'].median(), inplace=True)
print("Missing values after filling:")
print(iris_missing.isnull().sum())

# 10. Create new columns or modify existing ones.
iris['petal_ratio'] = iris['petal_length'] / iris['petal_width']
print("\nNew column 'petal_ratio' added:")
print(iris[['petal_length', 'petal_width', 'petal_ratio']].head())

# 11. Use groupby to aggregate data.
grouped = iris.groupby('species').mean()
print("\nGrouped average values by species:")
print(grouped)



# Step 5: Data Visualization with Matplotlib & Seaborn
# Objective: Visualize the data to identify patterns, trends, and insights.

# 12. Use Matplotlib to create a basic plot.
plt.figure(figsize=(8,4))
plt.plot(iris['sepal_length'], label='Sepal Length')
plt.title('Sepal Length Plot')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length (cm)')
plt.legend()
plt.show()

# 13. Create a histogram using Seaborn.
plt.figure(figsize=(8,4))
sns.histplot(iris['sepal_length'], kde=True, color='green')
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# 14. Plot a box plot for a clear view of data distribution.
plt.figure(figsize=(8,4))
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title('Sepal Length by Species')
plt.show()
```
