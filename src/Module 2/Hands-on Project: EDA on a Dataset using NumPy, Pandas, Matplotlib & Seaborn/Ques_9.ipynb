```python
# Title: Feature Engineering for Classification
# Description: Create a new feature that could help distinguish between species based on
# logical assumptions and verify its utility.
```

```python
import seaborn as sns
import pandas as pd

# Load Iris dataset
iris = sns.load_dataset('iris')

# Create a new binary feature based on logical assumption
iris['petal_longer_than_wide'] = iris['petal_length'] > iris['petal_width']

# Show how the new feature distributes across species
feature_summary = iris.groupby('species')['petal_longer_than_wide'].mean()
print(feature_summary)

# Optional: visualize the feature with respect to species
import matplotlib.pyplot as plt
sns.countplot(data=iris, x='species', hue='petal_longer_than_wide')
plt.title('Distribution of petal_longer_than_wide by species')
plt.show()
```
