```python
# Title: Correlation Analysis
# Description: Find the correlation between different features and visualize using a heatmap.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Select only numeric columns for correlation
numeric_iris = iris.select_dtypes(include='number')

# Compute the correlation matrix on numeric data only
corr_matrix = numeric_iris.corr()

# Display the correlation matrix
print("Correlation matrix:\n", corr_matrix)

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```
