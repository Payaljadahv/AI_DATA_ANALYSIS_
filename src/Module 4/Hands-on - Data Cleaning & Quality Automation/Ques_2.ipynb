```python
import pandas as pd

# Step 1: Load the Data (create sample data)
data = {
    'Name': ['Alice', 'Bob', None, 'David', 'Eve', 'Frank', 'Alice'],
    'Age': [25, None, 30, 22, None, 40, 25],
    'City': ['NY', 'LA', 'NY', None, 'LA', 'NY', 'NY']
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Step 2: Handling Missing Values

# 2.1 Identify Missing Values
print("\nMissing values count per column:")
print(df.isnull().sum())

# 2.2 Fill Missing Values
# Fill missing Age with the mean age
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing Name with 'Unknown'
df['Name'].fillna('Unknown', inplace=True)

# Fill missing City with mode (most frequent value)
df['City'].fillna(df['City'].mode()[0], inplace=True)

print("\nDataFrame after filling missing values:")
print(df)

# Step 3: Handling Duplicates

# 3.1 Identify duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())

# 3.2 Remove duplicates (keep first occurrence)
df_clean = df.drop_duplicates()

print("\nDataFrame after removing duplicates:")
print(df_clean)
```

```python
import pandas as pd

# --- Step 1: Standardize Text Data (Names to lowercase) ---
data = {
    'name': ['Alice', 'BOB', 'Charlie', 'diana'],
    'age': [23.4, 35.7, 29.9, 41.2]
}
df = pd.DataFrame(data)

# Convert 'name' to lowercase
df['name'] = df['name'].str.lower()

# --- Step 2: Format Numerical Data (Round age to nearest integer) ---
df['age'] = df['age'].round().astype(int)

print("After standardizing names and rounding age:")
print(df)
print("\n")

# --- Step 3: Combined Practice on Another Dataset ---
products = {
    'product': ['Apple', 'banana', 'ORANGE', 'mango'],
    'price': [0.99, 1.2345, 2.5, 3.456]
}
df_products = pd.DataFrame(products)

# Standardize product names to lowercase
df_products['product'] = df_products['product'].str.lower()

# Format prices to two decimal places
df_products['price'] = df_products['price'].round(2)

print("After standardizing product names and formatting prices:")
print(df_products)
```
