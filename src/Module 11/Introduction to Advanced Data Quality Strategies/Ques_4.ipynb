## AI & Machine Learning for Data Quality
**Description**: AI and machine learning can automate and enhance data quality checks by learning patterns and identifying anomalies more effectively than static rules.

**Task 1**: Training a model to predict and flag unusual trend patterns in sales data that
deviate from historical norms.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# 1. Create a mock time series sales dataset
np.random.seed(42)
date_range = pd.date_range(start='2023-01-01', periods=365, freq='D')
sales = np.random.normal(loc=200, scale=25, size=365)

# Inject some anomalies
sales[50] = 400   # spike
sales[120] = 50   # drop
sales[300] = 420  # spike

df = pd.DataFrame({'date': date_range, 'sales': sales})
df['day_of_year'] = df['date'].dt.dayofyear
df.set_index('date', inplace=True)

# 2. Train a model to identify anomalies (Isolation Forest)
model = IsolationForest(contamination=0.01, random_state=42)
df['anomaly'] = model.fit_predict(df[['sales', 'day_of_year']])
df['is_anomaly'] = df['anomaly'] == -1

# 3. Visualize the results
plt.figure(figsize=(14, 6))
plt.plot(df.index, df['sales'], label='Sales', color='blue')
plt.scatter(df[df['is_anomaly']].index, df[df['is_anomaly']]['sales'],
            color='red', label='Anomaly', marker='x')
plt.title('Sales Trend with Anomaly Detection')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.tight_layout()
plt.show()
```
