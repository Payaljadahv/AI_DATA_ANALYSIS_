```python
# Load data
import pandas as pd
from great_expectations.validator.validator import Validator
from great_expectations.data_context import get_context
from great_expectations.expectations.expectation import ExpectationConfiguration

context = get_context()

# Load your dataset
df = pd.read_csv("sample_data.csv")
batch_request = {
    "datasource_name": "default_pandas_datasource",
    "data_connector_name": "default_runtime_data_connector_name",
    "data_asset_name": "sample_data",  # Can be anything
    "runtime_parameters": {"batch_data": df},
    "batch_identifiers": {"default_identifier_name": "default"},
}

# Create validator
validator = context.get_validator(
    batch_request=batch_request,
    expectation_suite_name="my_suite",
)

# Add expectations
validator.expect_column_values_to_not_be_null("age")
validator.expect_column_values_to_be_between("age", min_value=0, max_value=100)
validator.expect_column_values_to_be_in_set("status", ["active", "inactive"])

# Save suite
validator.save_expectation_suite()
```

```python
import pandas as pd

# Sample customer data
df = pd.DataFrame({
    "customer_id": [1, 2, 3, 4, 5, 5],
    "name": ["Alice", "Bob", "Charlie", "David", "Eve", "Eve"],
    "email": ["alice@example.com", "bob@example.com", "charlie@example.com", "david@example.com", None, None],
    "age": [25, 17, 45, 62, 35, 35],
    "phone": ["1234567890", "0987654321", "123456789", "1112223333", "abcdefghij", "abcdefghij"]
})
print("üîç Null value summary:")
print(df.isnull().sum())

print("\nüìä Data types:")
print(df.dtypes)

print("\nüìà Descriptive statistics:")
print(df.describe())
duplicates = df[df.duplicated(subset=["customer_id", "email"], keep=False)]
print("‚ùå Duplicates found:")
print(duplicates)
invalid_age = df[(df["age"] < 18) | (df["age"] > 65)]
print("\n‚ùå Invalid age values:")
print(invalid_age)
import re

def is_valid_phone(val):
    return bool(re.fullmatch(r"\d{10}", str(val)))

df["phone_valid"] = df["phone"].apply(is_valid_phone)
invalid_phone = df[~df["phone_valid"]]
print("\n‚ùå Invalid phone numbers:")
print(invalid_phone[["phone"]])
def dq_score(col):
    total = len(col)
    nulls = col.isnull().sum()
    unique_ratio = col.nunique() / total
    score = 100 - (nulls / total) * 50 - (1 - unique_ratio) * 50
    return round(score, 2)

print("\n‚úÖ DQ Scores by Column:")
for column in df.columns[:-1]:
    print(f"{column}: {dq_score(df[column])}/100")
```
