```python
# Question: Evaluating Data Distribution
# Description: Analyze the distribution of a numeric column using histograms and descriptive statistics.

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, normaltest

# Load a sample dataset from seaborn
df = sns.load_dataset('tips')

# Analyze the 'total_bill' column
column = 'total_bill'

# Descriptive statistics
print("Descriptive Statistics:")
print(df[column].describe())
print("\nSkewness:", df[column].skew())
print("Kurtosis:", df[column].kurtosis())

# Histogram and KDE
plt.figure(figsize=(10, 6))
sns.histplot(df[column], bins=30, kde=True, color='skyblue')
plt.title(f'Distribution of {column}', fontsize=14)
plt.xlabel(column)
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()

# Normality tests
shapiro_stat, shapiro_p = shapiro(df[column])
dagostino_stat, dagostino_p = normaltest(df[column])

print("\nShapiro-Wilk Test:")
print(f"Statistic = {shapiro_stat:.4f}, p-value = {shapiro_p:.4f}")

print("\nD’Agostino and Pearson’s Test:")
print(f"Statistic = {dagostino_stat:.4f}, p-value = {dagostino_p:.4f}")

# Interpretation
print("\nInterpretation:")
if shapiro_p < 0.05 or dagostino_p < 0.05:
    print("The data is likely **not normally distributed** (p < 0.05).")
else:
    print("The data may be **normally distributed** (p ≥ 0.05).")
```
