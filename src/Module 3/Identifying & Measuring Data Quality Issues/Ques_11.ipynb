```python
# Common Data Errors Examples

# 1. Missing Data:
# Task 1: Review a dataset where some customer emails are missing. Identify how
# many records are incomplete.
# Task 2: Examine a sales dataset with missing transaction dates and determine the
# percentage of missing data.
# Task 3: Identify missing department information in an employee registry.






# 2. Duplicate Data:
# Task 1: Analyze a customer dataset with duplicate entries and count the number of
# duplicates.
# Task 2: Review supplier data and identify any repeated supplier names.
# Task 3: Examine a product inventory list for duplicates in product IDs.






# 3. Inconsistent Formatting:
# Task 1: Spot inconsistencies in date formats (e.g., DD/MM/YYYY vs. MM/DD/YYYY)
# in a dataset.
# Task 2: Identify phone numbers with varying formats in a contact list.
# Task 3: Review address data for discrepancies in state abbreviations (e.g., CA vs.
# Calif.).





# 4. Data Drift:
# Task 1: Compare monthly revenues over six months to identify data drift.
# Task 2: Analyze user engagement metrics from a web application over different
# quarters.
# Task 3: Review a stock price dataset to detect any anomalies over a year.



```

```python
import pandas as pd
import numpy as np
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------
# 1. Missing Data
# -------------------------
print("üîç MISSING DATA EXAMPLES")

# Task 1: Missing customer emails
df_customers = pd.DataFrame({
    'customer_id': range(1, 6),
    'email': ['john@example.com', None, 'jane@example.com', None, 'bob@example.com']
})
print("\nTask 1 - Missing Emails:\n", df_customers)
print("Missing email count:", df_customers['email'].isna().sum())

# Task 2: Missing transaction dates
df_sales = pd.DataFrame({
    'transaction_id': range(1, 6),
    'date': ['2023-01-01', None, '2023-01-03', None, '2023-01-05']
})
missing_pct = df_sales['date'].isna().mean() * 100
print("\nTask 2 - Missing Transaction Dates:\n", df_sales)
print(f"Missing transaction date percentage: {missing_pct:.2f}%")

# Task 3: Missing department info
df_employees = pd.DataFrame({
    'emp_id': [101, 102, 103, 104],
    'department': ['HR', None, 'IT', None]
})
print("\nTask 3 - Missing Departments:\n", df_employees)
print("Missing department count:", df_employees['department'].isna().sum())

# -------------------------
# 2. Duplicate Data
# -------------------------
print("\nüîç DUPLICATE DATA EXAMPLES")

# Task 1: Duplicate customers
df_customers_dup = pd.DataFrame({
    'customer_id': [1, 2, 2, 3, 4, 4, 4],
    'name': ['Alice', 'Bob', 'Bob', 'Carol', 'Dan', 'Dan', 'Dan']
})
print("\nTask 1 - Customer Duplicates:\n", df_customers_dup)
print("Duplicate rows:", df_customers_dup.duplicated().sum())

# Task 2: Repeated supplier names
df_suppliers = pd.DataFrame({
    'supplier_id': [1, 2, 3, 4, 5],
    'name': ['Alpha Inc.', 'Beta LLC', 'Alpha Inc.', 'Gamma Corp.', 'Beta LLC']
})
print("\nTask 2 - Repeated Supplier Names:\n", df_suppliers)
print("Duplicate supplier names:", df_suppliers.duplicated(subset='name').sum())

# Task 3: Duplicate product IDs
df_products = pd.DataFrame({
    'product_id': ['P001', 'P002', 'P003', 'P001', 'P004'],
    'product_name': ['Chair', 'Table', 'Lamp', 'Chair', 'Couch']
})
print("\nTask 3 - Duplicate Product IDs:\n", df_products)
print("Duplicate product IDs:", df_products.duplicated(subset='product_id').sum())

# -------------------------
# 3. Inconsistent Formatting
# -------------------------
print("\nüîç INCONSISTENT FORMATTING EXAMPLES")

# Task 1: Date formats
df_dates = pd.DataFrame({
    'raw_date': ['01/12/2023', '12/01/2023', '2023-01-12', '12-Jan-2023']
})
print("\nTask 1 - Inconsistent Date Formats:\n", df_dates)

# Task 2: Phone number formats
df_phones = pd.DataFrame({
    'phone': ['1234567890', '(123) 456-7890', '123-456-7890', '+1 123 456 7890']
})
print("\nTask 2 - Phone Number Formats:\n", df_phones)

# Task 3: State abbreviations
df_addresses = pd.DataFrame({
    'state': ['CA', 'Calif.', 'NY', 'New York', 'TX', 'Tex.']
})
print("\nTask 3 - Inconsistent State Abbreviations:\n", df_addresses)

# -------------------------
# 4. Data Drift
# -------------------------
print("\nüîç DATA DRIFT EXAMPLES")

# Task 1: Monthly revenue drift
months = pd.date_range(start='2024-01-01', periods=6, freq='M')
revenues = [10000, 10200, 9900, 9500, 8700, 12000]
df_revenue = pd.DataFrame({'month': months, 'revenue': revenues})
print("\nTask 1 - Monthly Revenue Drift:\n", df_revenue)
sns.lineplot(data=df_revenue, x='month', y='revenue', marker='o')
plt.title("Monthly Revenue Drift")
plt.grid(True)
plt.show()

# Task 2: User engagement drift
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
engagement = [300, 350, 280, 150]
df_engagement = pd.DataFrame({'quarter': quarters, 'active_users': engagement})
print("\nTask 2 - Engagement Drift:\n", df_engagement)
sns.barplot(data=df_engagement, x='quarter', y='active_users')
plt.title("Quarterly User Engagement")
plt.grid(True)
plt.show()

# Task 3: Stock price anomalies
dates = pd.date_range(start='2024-01-01', periods=12, freq='M')
prices = [100, 102, 98, 97, 150, 155, 160, 95, 93, 92, 180, 185]
df_stock = pd.DataFrame({'date': dates, 'price': prices})
print("\nTask 3 - Stock Price Drift:\n", df_stock)
sns.lineplot(data=df_stock, x='date', y='price', marker='o')
plt.title("Stock Price Over 12 Months")
plt.grid(True)
plt.show()
```
