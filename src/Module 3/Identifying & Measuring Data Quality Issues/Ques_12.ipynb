```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

# -------------------------------------
# 1. Descriptive Statistics
# -------------------------------------
print("üîç DESCRIPTIVE STATISTICS")

# Task 1: Sales figures - mean, median, mode
df_sales = pd.DataFrame({'sales': [100, 150, 200, 150, 300, 250, 150]})
mean_sales = df_sales['sales'].mean()
median_sales = df_sales['sales'].median()
mode_sales = df_sales['sales'].mode()[0]
print(f"Task 1 - Sales Figures: Mean={mean_sales}, Median={median_sales}, Mode={mode_sales}")

# Task 2: Customer ages - mean, median, mode
df_customers = pd.DataFrame({'age': [22, 25, 25, 30, 40, 25, 35]})
mean_age = df_customers['age'].mean()
median_age = df_customers['age'].median()
mode_age = df_customers['age'].mode()[0]
print(f"Task 2 - Customer Ages: Mean={mean_age}, Median={median_age}, Mode={mode_age}")

# Task 3: Website visits daily counts - mean, median, mode
df_visits = pd.DataFrame({'visits': [120, 150, 130, 150, 180, 150, 170]})
mean_visits = df_visits['visits'].mean()
median_visits = df_visits['visits'].median()
mode_visits = df_visits['visits'].mode()[0]
print(f"Task 3 - Daily Website Visits: Mean={mean_visits}, Median={median_visits}, Mode={mode_visits}")


# -------------------------------------
# 2. Distribution Analysis
# -------------------------------------
print("\nüîç DISTRIBUTION ANALYSIS")

# Task 1: Histogram of product prices
df_products = pd.DataFrame({'price': [10, 15, 10, 20, 30, 25, 15, 20, 10, 15]})
plt.figure(figsize=(6,4))
sns.histplot(df_products['price'], bins=5, kde=True)
plt.title("Task 1 - Distribution of Product Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

# Task 2: Distribution of exam scores
df_scores = pd.DataFrame({'score': [55, 65, 75, 85, 95, 65, 70, 80, 85, 90]})
plt.figure(figsize=(6,4))
sns.histplot(df_scores['score'], bins=6, kde=True)
plt.title("Task 2 - Distribution of Exam Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()

# Task 3: Distribution of order quantities
df_orders = pd.DataFrame({'quantity': [1, 2, 5, 3, 4, 2, 2, 1, 5, 3, 3]})
plt.figure(figsize=(6,4))
sns.histplot(df_orders['quantity'], bins=5, kde=False)
plt.title("Task 3 - Distribution of Order Quantities")
plt.xlabel("Quantity")
plt.ylabel("Count")
plt.show()


# -------------------------------------
# 3. Data Type Validation
# -------------------------------------
print("\nüîç DATA TYPE VALIDATION")

# Task 1: Validate numeric fields (sales)
df_numeric_check = pd.DataFrame({'sales': ['100', '150', 'abc', '200', 'NaN', '250']})
df_numeric_check['is_numeric'] = pd.to_numeric(df_numeric_check['sales'], errors='coerce').notna()
print("Task 1 - Numeric Field Validation:\n", df_numeric_check)

# Task 2: Validate date formats
df_dates = pd.DataFrame({'transaction_date': ['2023-01-01', '01/02/2023', '2023/03/01', 'invalid_date', '2023-05-01']})

def validate_date(date_str):
    for fmt in ('%Y-%m-%d', '%d/%m/%Y', '%Y/%m/%d'):
        try:
            datetime.strptime(date_str, fmt)
            return True
        except:
            continue
    return False

df_dates['valid_date'] = df_dates['transaction_date'].apply(validate_date)
print("\nTask 2 - Date Format Validation:\n", df_dates)

# Task 3: Validate email formats
df_emails = pd.DataFrame({'email': ['john@example.com', 'invalid-email', 'jane.doe@mail.com', 'user@site', 'user.name@domain.co']})

email_pattern = re.compile(r'^[\w\.-]+@[\w\.-]+\.\w+$')
df_emails['valid_email'] = df_emails['email'].apply(lambda x: bool(email_pattern.match(x)))
print("\nTask 3 - Email Format Validation:\n", df_emails)
```
