```python
import pandas as pd

# Part 1: Load a Dataset & Check Missing Values

# Task 1: Load Iris dataset (from sklearn for convenience)
from sklearn.datasets import load_iris

iris_sklearn = load_iris()
iris = pd.DataFrame(data=iris_sklearn.data, columns=iris_sklearn.feature_names)
iris['target'] = iris_sklearn.target  # Add target column for completeness

# Check for missing values in the entire dataset
missing_values = iris.isnull().sum()
print("Missing values in each column of Iris dataset:")
print(missing_values)


# Part 2: Identify Duplicates & Inconsistencies

# Task 2: Identify inconsistent entries in 'Gender' column

# Example dataset with inconsistent 'Gender' entries
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Gender': ['F', 'M', 'Male', 'female', 'male']
}
df_gender = pd.DataFrame(data)

print("\nOriginal 'Gender' column values:")
print(df_gender['Gender'])

# Normalize the Gender column by mapping all variants to standard values
gender_mapping = {
    'M': 'Male',
    'Male': 'Male',
    'male': 'Male',
    'F': 'Female',
    'Female': 'Female',
    'female': 'Female'
}

# Convert all values to lowercase for matching, then map
df_gender['Gender_Normalized'] = df_gender['Gender'].str.lower().map({
    'm': 'Male',
    'male': 'Male',
    'f': 'Female',
    'female': 'Female'
})

print("\nNormalized 'Gender' column values:")
print(df_gender[['Gender', 'Gender_Normalized']])


# Part 3: Generate a Data Quality Report

# Task 3: Generate basic descriptive statistics for the Iris dataset
print("\nIris Dataset Basic Descriptive Statistics:")
print(iris.describe())
```
