```python
import os
import pandas as pd

# Check current working directory
cwd = os.getcwd()
print("Current working directory:", cwd)

# List files in current directory
files = os.listdir()
print("Files in current directory:", files)

# Attempt to load customer_data.csv
file_name = 'customer_data.csv'

if file_name in files:
    try:
        df = pd.read_csv(file_name)
        print(f"\nSuccessfully loaded '{file_name}'")
        
        # Part 1: Check missing values in 'Email' and 'Phone'
        missing_email = df['Email'].isnull().sum()
        missing_phone = df['Phone'].isnull().sum()
        print(f"Missing values in 'Email': {missing_email}")
        print(f"Missing values in 'Phone': {missing_phone}")

        # Part 2: Identify duplicate emails
        duplicate_emails = df[df.duplicated(subset=['Email'], keep=False)]
        print(f"\nNumber of duplicate email entries: {duplicate_emails.shape[0]}")
        print("\nDuplicate email records:")
        print(duplicate_emails)

        # Part 3: Data Quality Report function
        def customer_data_quality_report(dataframe):
            print("\n--- Customer Data Quality Report ---")
            
            print("\nMissing values per column:")
            print(dataframe.isnull().sum())
            
            total_duplicates = dataframe.duplicated().sum()
            print(f"\nTotal duplicate rows in dataset: {total_duplicates}")
            
            duplicate_email_count = dataframe.duplicated(subset=['Email']).sum()
            print(f"Duplicate emails count: {duplicate_email_count}")
            
            print("\nBasic statistics (numeric columns):")
            print(dataframe.describe())
            
            print("\nBasic statistics (categorical columns):")
            print(dataframe.describe(include=['object', 'category']))

        customer_data_quality_report(df)

    except Exception as e:
        print(f"An error occurred while loading or processing the file: {e}")
else:
    print(f"File '{file_name}' not found in directory: {cwd}")
    print("Please upload or move the file to this directory.")



```
