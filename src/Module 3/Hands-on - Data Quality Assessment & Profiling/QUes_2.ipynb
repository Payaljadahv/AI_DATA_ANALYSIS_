```python
import pandas as pd

# Part 1: Load a Dataset & Check Missing Values

# Task 1: Load Titanic dataset
# If you don't have the dataset locally, you can load from seaborn as a quick example
import seaborn as sns
titanic = sns.load_dataset('titanic')

# Check missing values in 'age' column
missing_age_count = titanic['age'].isnull().sum()
print(f"Missing values in 'age' column: {missing_age_count}")



# Part 2: Identify Duplicates & Inconsistencies

# Task 2: Identify duplicate rows
duplicate_rows = titanic[titanic.duplicated()]
print(f"\nNumber of duplicate rows: {duplicate_rows.shape[0]}")



# Part 3: Generate a Data Quality Report

# Task 3: Summary report

def data_quality_report(df):
    print("\n--- Data Quality Report ---")
    # Missing values per column
    print("\nMissing values per column:")
    print(df.isnull().sum())
    
    # Duplicate rows count
    print(f"\nNumber of duplicate rows: {df.duplicated().sum()}")
    
    # Basic statistics summary for numerical columns
    print("\nBasic statistics (numerical columns):")
    print(df.describe())
    
    # Basic statistics summary for categorical columns
    print("\nBasic statistics (categorical columns):")
    print(df.describe(include=['object', 'category']))
    

data_quality_report(titanic)
```
