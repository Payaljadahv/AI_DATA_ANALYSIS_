```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Sample Titanic data
data = {'Sex': ['male', 'female', 'female', 'male', 'female']}
df = pd.DataFrame(data)

# Label Encoding
le = LabelEncoder()
df['Sex_LabelEncoded'] = le.fit_transform(df['Sex'])

# One-Hot Encoding
ohe = OneHotEncoder(sparse_output=False)
sex_ohe = ohe.fit_transform(df[['Sex']])
df_ohe = pd.DataFrame(sex_ohe, columns=ohe.categories_[0])

df = pd.concat([df, df_ohe], axis=1)

print(df)
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
X = iris.data
feature_names = iris.feature_names

# Min-Max Scaling first
mm_scaler = MinMaxScaler()
X_minmax = mm_scaler.fit_transform(X)

# Then Standardization on Min-Max scaled data
std_scaler = StandardScaler()
X_combined = std_scaler.fit_transform(X_minmax)

df = pd.DataFrame(X_combined, columns=feature_names)
print(df.head())
# Sample Titanic data
data = {'Sex': ['male', 'female', 'female', 'male', 'female'],
        'Embarked': ['S', 'C', 'Q', 'S', 'C']}
df = pd.DataFrame(data)

# One-Hot Encode 'Sex' and 'Embarked'
df_encoded = pd.get_dummies(df, columns=['Sex', 'Embarked'])
print(df_encoded)
from sklearn.preprocessing import OrdinalEncoder

data = {'Pclass': [3, 1, 2, 3, 1]}
df = pd.DataFrame(data)

# Define order: 1st class = highest (1), 3rd = lowest (3)
ordinal_encoder = OrdinalEncoder(categories=[[1, 2, 3]])
df['Pclass_ordinal'] = ordinal_encoder.fit_transform(df[['Pclass']])

print(df)

```
