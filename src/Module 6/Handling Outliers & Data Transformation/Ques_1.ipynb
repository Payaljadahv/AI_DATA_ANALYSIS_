```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example data - you can replace this with your own DataFrame and column
data = {
    'Monthly_Sales': [200, 220, 210, 230, 250, 270, 300, 4000, 280, 260, 255, 245]
}
df = pd.DataFrame(data)

# Plot boxplot for 'Monthly_Sales'
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Monthly_Sales'])
plt.title('Boxplot to Detect Outliers in Monthly_Sales')
plt.xlabel('Monthly Sales')
plt.show()
```
