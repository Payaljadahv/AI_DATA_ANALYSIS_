```python
# Question: Power Transformation (Box-Cox) Method
# Description: Use Box-Cox transformation to stabilize variance and make the data more normal.


```

```python
import pandas as pd
from scipy.stats import boxcox
import numpy as np
import matplotlib.pyplot as plt

# Sample positive data (replace with your dataset column)
data = {'Monthly_Sales': [200, 220, 210, 230, 250, 270, 300, 400, 280, 260, 255, 245]}
df = pd.DataFrame(data)

# Check data positivity (Box-Cox requires positive values)
if (df['Monthly_Sales'] <= 0).any():
    raise ValueError("Data must be positive for Box-Cox transformation")

# Apply Box-Cox transformation
df['Monthly_Sales_BoxCox'], fitted_lambda = boxcox(df['Monthly_Sales'])

print(f"Optimal lambda for Box-Cox: {fitted_lambda}")

# Plot original vs transformed data for comparison
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
plt.title('Original Data')
plt.hist(df['Monthly_Sales'], bins=10)
plt.subplot(1, 2, 2)
plt.title('Box-Cox Transformed Data')
plt.hist(df['Monthly_Sales_BoxCox'], bins=10)

plt.show()
```
